## Inference Clients

```python
from src.vllm_client import VllmClient, ArtClient, VllmRouter

inference_clients: list[VllmClient] = [ArtClient.from_art_model(model)]
vllm_server_ports = []
for port in vllm_server_ports:
    inference_clients.append(
        VllmClient.from_connection(
            port=port,
            base_model=model.base_model,
            model_name=model.get_inference_name(),
        )
    )

vllm_router: VllmRouter = VllmRouter(inference_clients)
```

## Training and Inference 

```python
from experiments.easy2hard.trainer import Easy2HardTrainer
from src.trainer import PromptConfig, DecompConfig, RolloutStage
from src.agent import BaseAgent, GuidedAgent
from src.configs import RolloutConfig
import random
import art

prompt_config = PromptConfig(
    system_root="dac_sys_prompt_gilad_v2_root",
    system_inter="dac_sys_prompt_gilad_v2_inter",
    system_leaf="dac_sys_prompt_gilad_v2_leaf",
    tasks_depleted="tasks_depleted_v2",
)

decomp_config = DecompConfig(
    max_depth=1,
    max_tasks=2,
    max_rounds=5,
)

class Easy2HardGuidedTrainer(Easy2HardTrainer):
    def create_agent(self, stage: RolloutStage) -> BaseAgent:
        client = self.vllm_router.next()
        decomp_config = self.decomp_config

        if stage == RolloutStage.Train and self.extra_config.get("randomize_decomp_depth", False):
            decomp_config = DecompConfig(
                max_depth=random.randint(0, decomp_config.max_depth or 10),
                max_tasks=decomp_config.max_tasks,
                max_rounds=decomp_config.max_rounds,
            )

        return GuidedAgent(
            model_name=self.model.get_inference_name(),
            openai_client=client.openai_client,
            prompt_config=self.prompt_config,
            decomp_config=decomp_config,
        )

rollout_config = RolloutConfig(
    kwargs={
        "max_completion_tokens": 5000,
        "temperature": 0.7,
        "top_p": 0.8,
        "verbose": True,
    }
)

trainer = Easy2HardGuidedTrainer(
    model=model,
    vllm_router=vllm_router,
    path_config=path_config,
    prompt_config=prompt_config,
    decomp_config=decomp_config,
    rollout_config=rollout_config,
)
```
